### Multiple Dispatch in Functions

```julia
f(x) = x^2
```

```julia
f(2)
```

```julia
f(x:: Int) = x^2

f(2)
```

```julia
f(x:: Float64) = x^2

f(2)
```

```julia
fun(x:: Int) = "x is an Integer"
fun(x:: Float64) = "x is an Float"
```

```julia
fun(2)
```

```julia
fun(2.)
```

```julia
fun(x:: Int) = "x is an Integer"
fun(x:: Float64) = "x is an Float"
fun(x:: Any) = "x is a generic object"
```

```julia
fun(2)
```

```julia
fun("Hello")
```

### BenchmarkTools

```julia
using BenchmarkTools
```

```julia
numbers = [i for i in 1:1000000];
```

```julia
function my_mean(numbers::Vector)
    s = 0
    N = size(numbers,1)
    for i in 1:N
        s += numbers[i]
    end
    return s/N
end



@btime my_mean(numbers)
```

```julia
function my_mean2(numbers::Vector)
    s = sum(numbers)
    N = size(numbers,1)
    return s/N
end



@btime my_mean(numbers)
```

```julia

@btime mean(numbers)
```
