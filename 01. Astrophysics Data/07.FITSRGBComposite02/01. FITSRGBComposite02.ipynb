### Summary

In this notebook we will use `astropy` to read a set of .fits files and display a composite version, based on information from different filters, of an astronomical image.

---

## 1. The .FITS Files

In this second example of image composition, we will use the .fits files of the galaxy NGC123 (M81) obtained from the SDSS database,

https://dr12.sdss.org/fields/name?name=M81

The images in bands *i*, *r* and *g* will be downloaded. For information on the bands in the SDSS see [here](https://www.sdss.org/instruments/camera/)

```python
import numpy as np
from matplotlib import pyplot as plt

from astropy.io import fits
from astropy.visualization import make_lupton_rgb

hduli = fits.open('data/NGC123i.fits')
hdulr = fits.open('data/NGC123r.fits')
hdulg = fits.open('data/NGC123g.fits')
hdulu = fits.open('data/NGC123u.fits')

i = hduli[0].data
r = hdulr[0].data
g = hdulg[0].data
u = hdulu[0].data
```

```python
u.shape
```

## 2. Image Display

```python
rgb_default = make_lupton_rgb(r, g, u)

plt.figure(figsize=(10,10))
plt.imshow(rgb_default, origin='lower')
plt.show()
```

```python
rgb_default = make_lupton_rgb(i, r, g, stretch=0.8, Q=18)

plt.figure(figsize=(10,10))
plt.imshow(rgb_default, origin='lower')
plt.show()
```
